```python
import tensorflow as tf
from tensorflow import keras
```

```python
VOCAB_SIZE = 10000 # needed for one hot encoding and embedding
MAX_SENT_LEN = 10 # needed for padding
EMBEDDING_DIM = 15
```

```python
# sentences
sentences = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understanding the meaning of words',
    'your videos are good'
]
```

onehot representation

```python
one_hot_reprs = [keras.preprocessing.text.one_hot(input_text=sent, n=VOCAB_SIZE, split=' ') for sent in sentences]
one_hot_reprs
```

```python
# padding
padded_docs = keras.preprocessing.sequence.pad_sequences(one_hot_reprs, padding='pre', maxlen=MAX_SENT_LEN)
padded_docs
```

Word Embedding Representation

```python
model = keras.Sequential([
    keras.layers.Embedding(VOCAB_SIZE, EMBEDDING_DIM, input_length=MAX_SENT_LEN)
])

model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
padded_docs[0]
```

```python
model.predict(padded_docs)[0] # 10 words * 15 embedding dims
```
